```python
#creating the filepath variable
filepath = '../input/housing-supply/3410013501-eng.csv'
```

```python
#read the csv file and skip the rows that are not needed(refer to the original csv file)
df = pd.read_csv(filepath,skiprows = 10)
#df = pd.read_csv(filepath,skiprows = -6)
# delete a row by index value 0 which is not necesaary for our case
df = df.drop(labels=0, axis=0)
```

```python
# These last rows are not necessary so we will drop these rows
#Number of rows to drop
n = 6
 
# Removing last n rows
df_dropped_last_n = df.iloc[:-n]
 
# Printing dataframe
df_dropped_last_n.tail()
```

```python
#We will first set the index to 'Type of unit' to find out whether it can give some insights
df_new = df_dropped_last_n.set_index("Type of unit")
df_new.head()
```

```python
#Fill NaN values with Housing estimates values 
df_new['Housing estimates'].fillna(method='ffill', inplace=True)
```

```python
#check whether we will fill NaN values with housing estimates values
df_new['Housing estimates'].unique()

```

```python
df_new = df_new.reset_index()
df_new.head()
```

```python
df_clean = df_new.drop(df_new[df_new['Type of unit'] == 'Total units'].index)
df_clean.head()
```

```python
df_clean.describe()
```

```python
df_clean.dtypes
```

```python
#need to replace comma if not then we cannot change object to integer
df_clean['Q1 2021'] = df_clean["Q1 2021"].str.replace(',','')
df_clean['Q2 2021'] = df_clean["Q2 2021"].str.replace(',','')
df_clean['Q3 2021'] = df_clean["Q3 2021"].str.replace(',','')
df_clean['Q4 2021'] = df_clean["Q4 2021"].str.replace(',','')
df_clean['Q1 2022'] = df_clean["Q1 2022"].str.replace(',','')
```

```python
df_clean['Q1 2021'] = df_clean['Q1 2021'].astype(int)
df_clean['Q2 2021'] = df_clean["Q2 2021"].astype(int)
df_clean['Q3 2021'] = df_clean["Q3 2021"].astype(int)
df_clean['Q4 2021'] = df_clean["Q4 2021"].astype(int)
df_clean['Q1 2022'] = df_clean["Q1 2022"].astype(int)
```
