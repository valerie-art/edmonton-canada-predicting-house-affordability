# 3. EDA on CPI 

```python
CPI_df = pd.read_csv('../input/housing-affordability-in-canada/CPI-inflation-by-region-1914-202.csv')
CPI_df.head()
```

```python
CPI_all_df = CPI_df[CPI_df['Products and product groups'] == 'All-items']
CPI_all_df.head()
```

```python
CPI_all_df.tail()
```

```python
# Return a Series containing counts of unique values in column UOM
CPI_all_df.UOM.value_counts()
```

Checked how many unique values are in UOM. Usually when comparing the CPI we have to use the same reference year.
Here only one unique value is present and the reference year is 2002 = 100, So we can use CPI data from 1914 - 2021 for comparing.

```python
fig = px.line(CPI_all_df, x="REF_DATE", y="CPI", color = 'GEO', title='Line chart for CPI in Canada')
fig.show()
```

After 1971 only provincial data is available. So plotting the CPI for year(1971 - 2021)

```python
fig = px.line( CPI_all_df[ (CPI_all_df.REF_DATE > 1970) & ( CPI_all_df['GEO'].isin(prov_terr) )], x="REF_DATE", y="CPI", color = 'GEO', title='Line chart for CPI in Canadian Provinces and territorries(1971-2021)')
fig.show()
```

```python
fig = px.line( CPI_all_df[ (CPI_all_df.REF_DATE > 1970) & ( CPI_all_df['GEO']=='Canada' )], x="REF_DATE", y="CPI", color = 'GEO', title='Line chart for CPI in Canada(1971-2021)')
fig.show()
```

```python
fig = px.bar( CPI_all_df[ (CPI_all_df.REF_DATE > 1970) & ( CPI_all_df['GEO'] =='Canada' )], x="REF_DATE", y="CPI", color = 'GEO', title='Bar chart for CPI in Canada(1971-2021)')
fig.show()
```

```python
fig = px.bar( CPI_all_df[ (CPI_all_df.REF_DATE > 2010) & ( CPI_all_df['GEO'].isin(prov_terr) )], x="REF_DATE", y="CPI", color = 'GEO', title='CPI in Canada(2010-2021)',barmode = 'group')
fig.show()
```

```python
fig = px.bar( CPI_df[ (CPI_df.REF_DATE == 2021) & ( CPI_df['GEO']=='Canada' )], x="Products and product groups", y="CPI", title='CPI for different products in Canada 2021', height = 1000)
fig.show()
```

### Top inflated products in Canada in 2021

```python
CPI_df[ (CPI_df.REF_DATE == 2021) & ( CPI_df['GEO']=='Canada' ) & ( CPI_df.CPI > 200 ) ].sort_values('CPI', ascending=False)
```
