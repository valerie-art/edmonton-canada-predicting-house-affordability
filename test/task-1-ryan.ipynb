```python
# mortgage rate data, changed column names and dropped empty columns
root = '/kaggle/input/mortgage-rate/'
mortgageData = pd.read_csv(os.path.join(root, '34100145.csv'))
mortgageData = mortgageData.drop(columns=['GEO', 'DGUID', 'UOM', 'UOM_ID', 'SCALAR_FACTOR', 'SCALAR_ID', 'VECTOR', 'COORDINATE', 'STATUS', 'SYMBOL', 'TERMINATED', 'DECIMALS'])
mortgageData = mortgageData.rename(columns={'REF_DATE': 'Date', 'VALUE': 'Mortgage Rate'})
```

```python
# interest rate data, changed column names
root = '/kaggle/input/interest-rate/'
interestData = pd.read_csv(os.path.join(root, 'IRSTCB01CAM156N.csv'))
interestData = interestData.rename(columns={'DATE': 'Date', 'IRSTCB01CAM156N': 'Interest Rate'})
```

```python
# merged dataframes together and aligned by date
rateData = pd.concat([interestData, mortgageData], axis=0)
rateData['Date'] = rateData['Date'].str[:7]
aggFunc = {'Mortgage Rate': 'sum', 'Interest Rate': 'sum'}
rateData = rateData.groupby(rateData['Date']).aggregate(aggFunc)
rateData = rateData.loc[:, (rateData != '0.00').any(axis=0)]
rateData
```
